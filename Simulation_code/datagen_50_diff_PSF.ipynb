```python
# Main code
Lambda = 0.520  # Wavelength in micrometers
NA = 0.04        # Numerical Aperture
P = 13.5          # Camera Pixel Size in micrometers
M = 1.5          # Magnification of optical setup

print("Pixel size of actual OBJ plane")
Obj_Pixel = P / M  # Pixel size of the real Obj plane
print(Obj_Pixel)

F0 = np.pi * M / P  # Frequency of the object sampling
Fcut = 2 * np.pi * NA / Lambda  # Cut-off Frequency of the optical system
N = 128


#screen=sio.loadmat("screen.mat")["screen"]
#PSF=screenPSF(N, NA, F0, screen)
#print(PSF.shape)
#PSF=PSF[0:256,0:256]
#sio.savemat("PSF.mat", {'PSF':PSF})
#PSF=sio.loadmat("PSF.mat")["PSF"][0:256,0:256]
```

```python
path='./cifar/raw/'#要改

#Train
output_speckle_train="./cifar/train/speckle/"
output_corr_train="./cifar/train/autocorr/"

#Test
#output_speckle_test="./cifar/Train/test/speckle/"
#output_corr_test="./cifar/Train/test/autocorr/"


path_list = os.listdir(path)
screen=sio.loadmat("screen.mat")["screen"]
PSF=screenPSF(N, NA, F0, screen)
PSF=PSF[0:256,0:256]
k=0
for item in path_list:
    if k%50==0:
        PSF=screenPSF(N, NA, F0, screen)#print(PSF.shape)
        PSF=PSF[0:256,0:256]
        print("PSF Updated!")
    impath=path+item
    print("开始处理"+impath)
    Obj1 = cv2.imread(path+item,0)
    Nx, Ny = Obj1.shape
    Obj = np.zeros(PSF.shape)
    Obj[N-Nx//2:N+Nx//2, N-Ny//2:N+Ny//2] = Obj1
    Img = conv2fft(PSF, Obj)


    cv2.imwrite("./cifar/train/gt/"+item[:-4]+".png",Obj1)
    sio.savemat(output_speckle_train+item[:-4]+".mat", {'data':Img})
    Img1=(Img-Img.min())/(Img.max()-Img.min())
    cv2.imwrite("./cifar/train/visual_speckle/"+item[:-4]+".png", Img1*255.0)
    Autocor_Img = autoCorr(Img)
    sio.savemat(output_corr_train+item[:-4]+".mat", {'data':Img})
    Autocor_Img1=(Autocor_Img-Autocor_Img.min())/(Autocor_Img.max()-Autocor_Img.min())
    cv2.imwrite("./cifar/train/visual_corr/"+item[:-4]+".png", Autocor_Img1*255.0)
        
    k=k+1


```
